### **Dimensionality Reduction with t-SNE**
 **Goal:** Use t-SNE to visualize high-dimensional data in 2D and understand group structures in complex datasets.

**Load or Simulate High-Dimensional Data**

```python
from sklearn.datasets import load_digits
import pandas as pd
import numpy as np

# Load dataset (64-dimensional features from 8x8 digit images)
digits = load_digits()
X = digits.data        # shape = (1797, 64)
y = digits.target      # digit labels (0â€“9)

print("Shape of X:", X.shape)
```

**Apply t-SNE for Dimensionality Reduction**

```python
from sklearn.manifold import TSNE

# Apply t-SNE to reduce to 2D
tsne = TSNE(n_components=2, perplexity=30, n_iter=1000, random_state=42)
X_tsne = tsne.fit_transform(X)

print("Shape after t-SNE:", X_tsne.shape)
```

**Visualize 2D Embeddings**

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a DataFrame for plotting
df_tsne = pd.DataFrame({
    'x': X_tsne[:, 0],
    'y': X_tsne[:, 1],
    'label': y
})

# Scatter plot colored by digit
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_tsne, x='x', y='y', hue='label', palette='tab10', s=50)
plt.title("t-SNE Visualization of Digit Embeddings (2D)")
plt.legend(title="Digit")
plt.grid(True)
plt.show()
```

**(Optional): 3D Visualization with t-SNE**

```python
from mpl_toolkits.mplot3d import Axes3D

# Run t-SNE for 3D
tsne_3d = TSNE(n_components=3, perplexity=30, n_iter=1000, random_state=42)
X_3d = tsne_3d.fit_transform(X)

# Plot in 3D
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(X_3d[:, 0], X_3d[:, 1], X_3d[:, 2], c=y, cmap='tab10', s=50)
ax.set_title("t-SNE 3D Visualization")
legend = ax.legend(*scatter.legend_elements(), title="Digit")
ax.add_artist(legend)
plt.show()
```
