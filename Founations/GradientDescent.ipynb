##**Build Your Own Gradient Descent from Scratch**

**Goal:** Use Python to implement linear regression with gradient descent and fit a line to
 synthetic data.

**Importing libraries**

```python
 import numpy as np
 import matplotlib.pyplot as plt
```

# **Generate Synthetic Data**

```python
# Create synthetic data: y = 3x + 4 + noise

np.random.seed(42)  # For reproducibility

X = 2 * np.random.rand(100, 1)  # 100 values between 0 and 2

y = 3 * X + 4 + np.random.randn(100, 1)  # Add noise

# Plot the data
plt.scatter(X, y)
plt.title("Synthetic Linear Data")
plt.xlabel("X")
plt.ylabel("y")
plt.show()
```

# **Initialize Parameters**

```python
# Initialize weights (slope) and bias (intercept)
m = 0  # Initial slope
b = 0  # Initial intercept

# Set hyperparameters
learning_rate = 0.05
epochs = 1000  # Number of iterations
# Number of training examples
n = len(X)
```

# **Implement Gradient Descent Loop**

```python
# Track loss over time
loss_history = []

# Gradient Descent loop
for i in range(epochs):

  # Predict current values
  y_pred = m * X + b

  # Compute gradients
  error = y_pred - y
  dm = (2/n) * np.sum(X * error)
  db = (2/n) * np.sum(error)

  # Update parameters
  m -= learning_rate * dm
  b -= learning_rate * db
  # Calculate and store loss (MSE)

  loss = (1/n) * np.sum(error**2)
  loss_history.append(loss)
  # Print updates every 100 iterations

  if i % 100 == 0:
      print(f"Epoch {i}: Loss = {loss:.4f}, m = {m:.4f}, b = {b:.4f}")
```

# **Visualize the Results**

```python
# Final fitted line
plt.scatter(X, y, label='Data')
plt.plot(X, m * X + b, color='red', label='Fitted Line')
plt.title("Linear Regression via Gradient Descent")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.show()
# Plot loss over epochs
plt.plot(range(epochs), loss_history)
plt.title("Loss over Epochs")
plt.xlabel("Epoch")
plt.ylabel("MSE Loss")
plt.grid(True)
plt.show()
```

##**Summary:**
**In this project**, we implemented Linear Regression from scratch using Gradient Descent. We created synthetic data based on a linear function with added noise, then trained a model to learn the best-fitting line.

**Throughout the training**, we updated the slope (m) and intercept (b) to minimize Mean Squared Error (MSE). Finally, we visualized both the fitted line and the loss over epochs. This project helped us understand the core idea behind linear regression and how optimization via gradient descent works.
