##**PCA for Dimensionality Reduction**

**Goal:** Use PCA to reduce the 64-dimensional handwritten digit data to 2D, and visualize
 class separability.

**Install Required Libraries**

```python
from sklearn.datasets import load_digits
from sklearn.decomposition import PCA
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

**Load the Digits Dataset**

```python
# Load the digits dataset (1797 samples, 64 features)
digits = load_digits()
X = digits.data
y = digits.target
print("Original shape of data:", X.shape)
```

**Apply PCA**

```python
# Reduce to 2 dimensions for visualization
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)
print("Reduced shape of data:", X_pca.shape)
```

**Visualize in 2D**

```python
# Convert to DataFrame for plotting
pca_df = pd.DataFrame(data=X_pca, columns=['PC1', 'PC2'])
pca_df['Label'] = y

# Plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=pca_df, x='PC1', y='PC2', hue='Label', palette='tab10', legend='full')
plt.title('PCA Projection of Digits Dataset (2D)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend(title='Digit Label')
plt.grid(True)
plt.show()
```

**Explained Variance**

```python
# Show how much variance is explained by each principal component
print("Explained variance ratio (first 2 components):", pca.explained_variance_ratio_)

# Plot cumulative explained variance for all components
pca_full = PCA().fit(X)
cumulative_variance = pca_full.explained_variance_ratio_.cumsum()

plt.figure(figsize=(8, 4))
plt.plot(range(1, len(cumulative_variance)+1), cumulative_variance, marker='o')
plt.title("Cumulative Explained Variance by PCA Components")
plt.xlabel("Number of Components")
plt.ylabel("Cumulative Explained Variance")
plt.axhline(y=0.95, color='r', linestyle='--', label='95% threshold')
plt.grid(True)
plt.legend()
plt.show()
```

## **Summary:**

**This project** demonstrates how to use Principal Component Analysis (PCA) to reduce the dimensionality of the handwritten digits dataset from 64 features to 2 for visualization.

**The first** two principal components capture ~28.5% of the total variance. A 2D scatter plot was generated to visualize how digits are distributed in the new space.

**Additionally**, we plotted the cumulative explained variance to understand how many components are needed to retain 95% of the data's information  approximately 40 components are sufficient. PCA is an effective method for compressing data, visualizing patterns, and preparing data for further machine learning tasks.
